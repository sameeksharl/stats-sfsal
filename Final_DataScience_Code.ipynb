# Data Cleaning

```python
#Claculating the NaN Percentage
missing_per = df.isnull().sum()*100/len(df)
missing_per
```

```python
# Removing columns Notes and Status as it contains 70%+ NaN values
df.drop("Notes", axis = 1, inplace = True)
df.drop("Status", axis = 1, inplace = True)
```

```python
#Data Frame after Droping
df
```

```python
#Convert column to a numeric type.
#Strings will be set as NaN.
df.BasePay = pd.to_numeric(df.BasePay, errors="coerce")
df.OvertimePay = pd.to_numeric(df.OvertimePay, errors="coerce")
df.OtherPay = pd.to_numeric(df.OtherPay, errors="coerce")
df.Benefits = pd.to_numeric(df.Benefits, errors="coerce")
df.TotalPay = pd.to_numeric(df.TotalPay, errors="coerce")
df.TotalPayBenefits = pd.to_numeric(df.TotalPayBenefits, errors="coerce")
```

```python
#Rows that contain Not provided as EmployeeName and Job Title
df[df.EmployeeName.str.contains('Not provided')]
```

```python
#Since only 4 rows contain Not provided we will drop these Rows
df = df[~df.EmployeeName.str.contains('Not provided')]
```

```python
#Data Frame after removing rows containing Not provided
df
```

```python
missing_per = df.isnull().sum()*100/len(df)
missing_per
```

```python
#Converting NaN values to inf as next operation will remove the rows containing NaN
df = df.fillna({'Benefits':float('inf'), 'BasePay':float('inf')})
```

```python
df.head()
```

```python
#Including all rows containing positive values (NaN's will not be considered so for this reason the NaN's was changed to inf)
df = df[df['BasePay']>=0]
df = df[df['OvertimePay']>=0]
df = df[df['OtherPay']>=0]
df = df[df['Benefits']>=0]
df = df[df['TotalPay']>=0]
df = df[df['TotalPayBenefits']>=0]
```

```python
df
```

```python
#Replacing inf back to NaN values to fill in later
df = df.replace(float('inf'), np.nan)
```

```python
df
```

```python
#Replacing NaN Values of Benefits and BasePay according to mean grouped by job title
df.groupby("JobTitle").mean()
```

```python
#Replacing NaN with mean (grouped by JobTitle)
df['BasePay'] = df['BasePay'].fillna(df.groupby("JobTitle")['BasePay'].transform('mean'))
df['Benefits'] = df['Benefits'].fillna(df.groupby("JobTitle")['Benefits'].transform('mean'))
```

```python
#Since mean Benefits of some JobTitles is NaN We replace them with 0
df = df.fillna({'Benefits':0})
```

```python
df
```

```python
missing_per = df.isnull().sum()*100/len(df)
missing_per
```

```python
#Since BasePay Still has few(4) NaN values we will remove them
df[df["BasePay"].isnull()]
```

```python
# Since count of all the below JobTitles are only one we can remove them
df[df['JobTitle'] == 'Captain, (Fire Department)']
```

```python
df[df['JobTitle'] == 'Public Safety Comm Tech']
```

```python
df[df['JobTitle'] == 'Drug Court Coordinator']
```

```python
df[df['JobTitle'] == 'IS Technician Assistant']
```

```python
# Removing the rows
df = df[~df["BasePay"].isnull()]
```

```python
#All the NaN has been removed
missing_per = df.isnull().sum()*100/len(df)
missing_per
```

```python
df
```

```python
# Formating  EmployeeName, JobTitle and Agency
df.loc[:, 'EmployeeName'] = df.loc[:, 'EmployeeName'].apply(lambda x: string.capwords(x))
df.loc[:, 'JobTitle'] = df.loc[:, 'JobTitle'].apply(lambda x: string.capwords(x))
df.loc[:, 'Agency'] = df.loc[:, 'Agency'].apply(lambda x: string.capwords(x))
```

# Hypothesis Testing

```python
import statsmodels.api as sm
```

```python
#Taking NULL hypothesis as the mean of the column grouped by JobTitle
a=df.groupby('JobTitle').mean()
a
```

```python
j1=df[df["JobTitle"]=='Elections Clerk']

j2=df[df["JobTitle"]=='Patient Care Assistant']

j3=df[df["JobTitle"]=='Senior Eligibility Worker']
```

```python
sm.stats.ztest(j1["BasePay"], value = a["BasePay"]['Elections Clerk'], alternative = "larger")
```

```python
sm.stats.ztest(j2["BasePay"], value = a["BasePay"]['Patient Care Assistant'], alternative = "two-sided")
```

```python
sm.stats.ztest(j3["TotalPay"], value = a["TotalPay"]['Senior Eligibility Worker'], alternative = "smaller")
```
